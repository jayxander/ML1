```python
import itertools as it
from itertools import groupby
import collections

def avgby(tbl, col, by):
    # get each column
    def get_nums(num):
        return [row[num] for row in tbl]

    col_nums = get_nums(col)
    col_group = get_nums(by)

    keys_and_values = list(zip(col_group, col_nums))

    list_a = []
    for key, group in groupby(keys_and_values, lambda x: x[0]):
        list_a += [[key, [g[1] for g in group]]]

    # creates dictionary from list
    dictionary = { value[0]: value[1] for value in list_a }

    for key in dictionary:
        avg = sum(dictionary[key]) / len(dictionary[key])
        dictionary[key] = avg
    
    return dictionary

test = [[12,  8  ,1, 11  ,5],
        [15,  0  ,0 , 1 ,12],
        [ 7 ,13  ,0  ,6 ,25],
        [18 ,20 , 1 ,18 ,20],
        [11 ,28,  0, 28, 29]]

avgby(test, 4, 2)

# SORT IT OUT


# Calculating the average of col 4, grouped by col 2
# OUT ╦ {1: 20.0, 0: 29.0}
#     ╚ ERROR: Expected:  {0: 22.0, 1: 12.5}
```
