```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
# list of column names
ufo.columns


# RENAME COLUMN => looks like dictionary
# inplace means operation occurs inplace, affect underlying dataframe
ufo.rename(columns = {'Colors Reported': 'Colors_Reported', 'Shape Reported': 'Shape_Reported'}, inplace=True)
ufo.columns


# RENAME COLUMN => simpler and overwrites the column names
ufo_cols = ['city', 'colors reported', 'shape reported', 'state', 'time']
ufo.columns = ufo_cols


# RENAME COLUMN => quickest method => replaces first argument with second (spaces with underscores)
ufo.columns = ufo.columns.str.replace(' ', '_')

ufo.head()
```

```python
# overwrite column names while reading the file
# header=0 means the zeroth row of the dataframe has existing column names to overwrite
ufo = pd.read_csv('http://bit.ly/uforeports', names=ufo_cols, header=0)
ufo.head()
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
# sorts by alphabetical order and returns SERIES
movies.title.sort_values()
# OR
movies['title'].sort_values()

# reverse alphabetical order with parameter False
movies.title.sort_values(ascending=False)


# DOES NOT AFFECT UNDERLYING DATA
```

```python
# sort DATAFRAME BY SERIES

movies.sort_values('title')
movies.sort_values('duration', ascending=False)
```

```python
# sort by more than one column => first sorts by first parameter, then by second
# passes a list of strings, instead of single string

movies.sort_values(['content_rating', 'duration'])
```

```python
type(False)

# create a list of booleans that tell us which rows match a given condition
# which rows match the given condition (>= 200 minutes in duration)
# and returns only the matching rows

booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)

        
# also does the same thing as above      
is_long = movies.duration >= 200

is_long
```

```python
booleans[0:5]
```

```python
len(booleans)
```

```python
# creates booleans into a pandas series
is_long = pd.Series(booleans)
is_long.head()
```

```python
# shows the new dataframe with only matching true values
movies[is_long]
```
